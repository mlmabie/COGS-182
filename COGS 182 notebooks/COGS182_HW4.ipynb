The Bellman Optimality Equations:
$$
v_*(s) = \max_a\sum_{s',r}p(s',r|s,a)\left[r+\gamma v_*(s')\right]\\
q_*(s,a) = \sum_{s',r}p(s',r|s,a)\left[r+\gamma \max_{a'}q_*(s',a')\right]\\
$$
Assuming knowledge of the environment's dynamics, this allows you to express the value of each state as a function of the values of its adjacent states. To factor policy, multiply by $\sum_a\pi(a|s)$.

### 3. (2 points) What do the final values in the grid world represent? Hint: How does reward relate to steps taken?

The final values in the grid world represent the penalties incentivizing the fewest steps to terminal state.

For any initial starting position, the smallest penalizing choice is balanced relative to every next step's choice, creating an optimal path.

```
import numpy as np
import pandas as pd

pd.set_option("display.precision", 2)

# recreating original reward function
v = np.zeros(16)
display(pd.DataFrame(np.reshape(v,(-1,4))))
for i in range(140):
  vn = np.zeros(16)
  vn[1] = -1 +.25*(v[1] + v[2] + v[5])
  vn[2] = -1 + .25*(v[1] + v[2] + v[3] + v[6])
  vn[3] = -1 + .25*(v[2] + 2*v[3] + v[7])
  vn[4] = -1 + .25*(v[4] + v[5] + v[8])
  vn[5] = -1 + .25*(v[1] + v[4] + v[6] + v[9])
  vn[6] = -1 + .25*(v[2] + v[5] + v[7] + v[10])
  vn[7] = -1 + .25*(v[3] + v[6] + v[7] + v[11])
  vn[8] = -1 + .25*(v[4] + v[8] + v[9] + v[12])
  vn[9] = -1 + .25*(v[5] + v[8] + v[10] + v[13])
  vn[10] = -1 + .25*(v[6] + v[9] + v[11] + v[14])
  vn[11] = -1 + .25*(v[7] + v[10] + v[11])
  vn[12] = -1 + .25*(v[8] + 2*v[12] + v[13])
  vn[13] = -1 + .25*(v[9] + v[12] + v[13] + v[14])
  vn[14] = -1 + .25*(v[10] + v[13] + v[14])
  v = vn
display(pd.DataFrame(np.reshape(v,(-1,4))))
```
